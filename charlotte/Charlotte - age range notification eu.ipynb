```python
import pandas as pd 
import sklearn
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('2_ecdc/agerangenotificationeu.csv')
```

```python
df.head()
```

```python
country_df = df.drop(["country_code", "source"], axis=1)
#country_df.set_index("country")

country_df["rate_14_day_per_100k"] = country_df["rate_14_day_per_100k"].fillna(0)
country_df.head()
```

```python
## week 36 31/08/20 to week 48 29/11/20
country_df['week_number'] = df.year_week.str[-2:]
country_df["week_number"] = pd.to_numeric(country_df["week_number"])

mask = (country_df['week_number'] >= 36) & (country_df['week_number'] < 49)
europe_2_wave_df = country_df.loc[mask] 

country_df = europe_2_wave_df.drop(["year_week"], axis=1)
```

```python
austria_df = country_df[country_df["country"]== "Austria"].copy()
austria_df = austria_df.drop(["country", "population"], axis=1)

austria_df = pd.melt(austria_df,
                    id_vars=["week_number", "age_group"],
                    var_name="Stat") 
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x="Stat", y="value",
             data=austria_df,
             hue="age_group",
             dodge=True) # separate points by hue

plt.title("Austria")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.figure(figsize=(10,6))
sns.swarmplot(x="Stat", y="value",
             data=austria_df,
             hue="week_number",
             dodge=True) # separate points by hue

plt.title("Austria")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
df = pd.read_csv('2_ecdc/testing.csv')
df.head()
```

```python
country_df = df.drop(["country_code", "testing_data_source"], axis=1)

## week 36 31/08/20 to week 48 29/11/20
country_df['week_number'] = df.year_week.str[-2:]
country_df["week_number"] = pd.to_numeric(country_df["week_number"])

mask = (country_df['week_number'] >= 36) & (country_df['week_number'] < 49)
europe_2_wave_df = country_df.loc[mask] 

country_df = europe_2_wave_df.drop(["year_week"], axis=1)


country_df.head()
```

```python
outlier_countries = ["Iceland", "Andorra", "San Marino"]

country_df = country_df[~country_df['country'].isin(outlier_countries)]
```

```python
country_df = country_df.set_index(["week_number","country"])

country_df.head()
```

```python
tests_df = country_df.testing_rate ## tests per million
tests_df = tests_df.unstack()
tests_df.plot(figsize=(10,10))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
tests_df = country_df.positivity_rate 
tests_df = tests_df.unstack()
tests_df.plot(figsize=(10,10))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
